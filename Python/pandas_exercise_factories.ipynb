## A 공장

```python
"""
A공장의 2019.01.01부터 10일간 생산량을 Series로 저장
생산량의 평균이 50이고 표준편차가 5인 정규분포에서
Random하게 추출(생산량은 정수로 처리)

예:
2019-01-01 54
2019-01-01 49
...
2019-01-11 77

B공장은 평균70표준편차9 정규분포 난수


인덱스를 날짜로 잡고, value가 생산량
** 생산량은 (50,5)의 정규분포 난수 발생

"""

# 1. 날짜 값 갖고 오는거
# 2. 정규분포 난수 발생하는거
# 3. 판다스 배열에 Values 저장
# 4. 판다스 배열에 인덱스(날짜) 지정

from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("A공장 생산률")

np.random.seed(0)

'''
1. 날짜 생성 및 더하기
'''
today = date.today()
# timedelta값은 시간 간격. 오늘날짜에 n일 수를 빼고 더할수 있다.
months = relativedelta(months=-1)
days = timedelta(days=-18)
today = today+days+months
print(today)

day_index = []

for d in range(0,10):
    days = timedelta(days=d)
    daily = today+days
    day_index.append(daily)

# print(day_index)
    
'''
2. 정규분포 난수 발생 (50,5)
'''
# 정규분포>난수값을 추출해서 array를 생성
# 추출된 난수값들이 정규분포를 띈다?

mean = 50 # 평균
std = 5 # 표준편차
arr = np.random.normal(mean,std,(10))
# 평균이 0이고 표준편차가 1인 10개(1차원) 난수 생성
# print(arr)
plt.plot(arr)
plt.axis([0,9,0,100])
plt.title("A factory production", fontsize=24)
plt.xlabel("10 day period",fontsize=18)
plt.ylabel("Production rate",fontsize=18)
plt.show()

'''
3. 판다스 배열에 values 저장
4. 판다스 배열에 인덱스(날짜) 지정
'''
A_co = pd.Series(arr, index=day_index)
print(A_co)
```

## A,B공장 한꺼번에!

```python
from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

'''
1. 날짜 생성 및 더하기
'''

# 반복문 돌려서 날짜 뽑기
start_day = datetime(2019,1,1)
s_date = [(start_day + timedelta(days=x)) for x in range(10)]
print(s_date[0])

# 판다 date_range기능 써서 날짜 뽑기
dti = pd.Series(pd.date_range('2019-01-01', freq='D', periods=10))
print(dti[0])
    
'''
2. 정규분포 난수 발생
50,5
70,9
'''
np.random.seed(0)

# 정규분포>난수값을 추출해서 array를 생성
# 추출된 난수값들이 정규분포를 띈다?

mean = 50 # 평균
std = 5 # 표준편차
arr1 = np.random.normal(mean,std,(10))
# 평균이 0이고 표준편차가 1인 10개(1차원) 난수 생성
# print(arr)

np.random.seed(1)
mean = 70
std = 9
arr2 = np.random.normal(mean,std,(10))

'''
3. 판다스 배열에 values 저장
4. 판다스 배열에 인덱스(날짜) 지정
'''
# 판다 date 사용해서 시리즈 생성
# A_co = pd.Series(arr1, index=dti)
data_a = {
    "Date":dti,
    "Rate":arr1
}
A_co = pd.DataFrame(data_a)
# 반복문 date 돌려서 시리즈 생성
# 이번엔 반복문이 더 빠르다;; 2배 정도 빠름..
# B_co = pd.Series(arr2, index=s_date)
data_b = {
    "Date":s_date,
    "Rate":arr2
}
B_co = pd.DataFrame(data_b)

companies = pd.merge(A_co,B_co, on="Date",how="inner")
display(companies)

'''
graph PLOT
'''
plt.plot(companies["Rate_x"], label="A co.")
plt.plot(companies["Rate_y"], label="B co.")
plt.axis([0,9,0,100])
plt.title("Factory Production", fontsize=24)
plt.xlabel("10 day period",fontsize=18)
plt.ylabel("Production rate",fontsize=18)
plt.legend()
```

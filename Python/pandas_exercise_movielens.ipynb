### 2. 각 사용자별 평균 평점

#### 사용자1번의 평균 평점

```python
# ratings파일 기반으로 users라는 사용자 목록 생성
users = ratings["userId"].unique()
# 사용자목록 0번 (사용자1번)의 마스크 생성
user1_mask = ratings['userId']==users[0]
# 사용자1번 마스크 사용해서 사용자1번의 리뷰자료들 출력 > 사용자1 리스트 생성
user1 = ratings.loc[user1_mask]
# 사용자1 리스트 기반으로 평균 평점 뽑음
user1['rating'].mean()
```

#### 각 사용자의 평균 평점

```python
# ratings파일 기반으로 users라는 사용자 목록 생성
users = ratings["userId"].unique()

user_mean = []

for user in users:
    mask = ratings['userId']==user
    user_prf = ratings.loc[mask]
    user_mean.append([user, user_prf['rating'].mean()])

user_mean = pd.DataFrame(user_mean,columns=["사용자","평균 평점"])
user_mean.head()
```

### 3. 각 영화별 평균 평점

#### 영화 1번 제목, 평균 평점 출력

```python
# 여기선 ratings, movie_list 둘다 사용해야 된다.

movies = ratings["movieId"].unique()
mov1_mask = ratings["movieId"]==movies[0]
mov1_rates = ratings.loc[mov1_mask]
mov1_mean = mov1_rates["rating"].mean() # 3.92~ 나온다
mov1_min = mov1_rates["rating"].min()
mov1_max = mov1_rates["rating"].max()

movie_list["min"] = np.nan
movie_list["max"] = np.nan
movie_list["mean"] = np.nan

movie_list.loc[ movie_list["movieId"]==movies[0],"min" ]= mov1_min
movie_list.loc[ movie_list["movieId"]==movies[0],"max" ]= mov1_max
movie_list.loc[ movie_list["movieId"]==movies[0],"mean" ]= mov1_mean

movie_list.head()
```

#### 모든 영화의 min, max, mean값 부여!

```python
%%time

movies = ratings["movieId"].unique()

for movie in movies:
    _mask = ratings["movieId"]==movie
    _rates = ratings.loc[_mask]
    _mean = _rates["rating"].mean()
    _min = _rates["rating"].min()
    _max = _rates["rating"].max()
    movie_list.loc[ movie_list["movieId"]==movie, "min" ] = _min
    movie_list.loc[ movie_list["movieId"]==movie, "mean" ] = _mean
    movie_list.loc[ movie_list["movieId"]==movie, "max" ] = _max
    
print("Success!")
```

```python
display(movie_list.head())
display(movie_list.tail())
```

### 4. 가장 높은 평균 평점 기준으로 영화 출력

```python
h_rate = movie_list["mean"].max()

# movie_list["reviewed"] = np.nan

movie_list.loc[ movie_list["mean"]==h_rate ]
```

### - 리뷰한 숫자 추가!

```python
%%time

reviewed = ratings["movieId"].value_counts()
reviewed[1]

for movie in movies:
    movie_list.loc[ movie_list["movieId"]==movie, "reviewed" ] = reviewed[movie]
```

```python
# movie_list.sort_values(by=['reviewed'], inplace=True, ascending=False)

movie_list.loc[ movie_list["mean"]==h_rate ].sort_values(by='reviewed', ascending=False).head()
```

```python
movie_list.loc[ movie_list["mean"]> 4.5 ].sort_values(by='reviewed', ascending=False).head()
```

### 5. comedy 영화 중 평점이 가장 낮은 영화의 제목

#### Series.str.contains() 이 함수를 사용해서 문자열에 있는 값에 대한 boolean mask가 가능!

```python
comedy_mask = movie_list.loc[:,"genres"].str.contains("Comedy")
movie_list[ comedy_mask ].head()
```

#### 코미디 영화의 가장 낮은 평균 평점

```python
low_comedy = movie_list[ comedy_mask ][ "mean" ].min()
low_comedy
```

```python
comedy_list = movie_list[ comedy_mask ]
comedy_list[ movie_list["mean"]== low_comedy].head()
```

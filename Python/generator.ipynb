# Make a Generator

```python
import pandas as pd
import numpy as np

df = pd.DataFrame([], columns=list(range(10)), index=list(range(100)))
df.fillna("x", inplace=True)
df[2] = list(range(100))

def butcher(dataset, batch_size):
    for i in range(0, dataset.shape[0], batch_size):
        yield dataset.iloc[i:i+batch_size]

batch = butcher(df, 5)

print(type(batch))

display(next(batch), next(batch), next(batch))
```

### inline 형태로 출력
numpy 배열, line graph 출력

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# matplotlib의 기본설정
# 1. graphc ruler(격자)의 사용여부
plt.rcParams["axes.grid"] = True
# 2. x,y 크기 지정
plt.rcParams["figure.figsize"] = (8,4)

# plot할 데이터 준비
x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum() # -100~100사이 50개 난수 생성
# cumulative summation
# 원소1,2를 더해서, 결과치로 원소2번으로 갈아침. 리스트 끝까지 반복.

# 그래프 label 지정
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")

plt.plot(x_data, y_data,"r") # "r"은 red 색 지정
plt.show()
```

## label 한글 encoding 처리

### font_manager 사용

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 사용할 font의 FontProperties
font_path = "C:\Windows\Fonts/malgun.ttf"
font_prop = fm.FontProperties(fname=font_path, size=14)

plt.rcParams["axes.grid"] = True
plt.rcParams["figure.figsize"] = (8,4)
x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum() # -100~100사이 50개 난수 생성
plt.title("가격변동 추이", fontproperties=font_prop)
plt.xlabel("기간(일)", fontproperties=font_prop)
plt.ylabel("가격", fontproperties=font_prop)
plt.plot(x_data, y_data,"r") # "r"은 red 색 지정
plt.show()
```

## 그래프 기본 설정

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams["axes.grid"] = True
plt.rcParams["figure.figsize"] = (8,4)
# label에 minus기호를 사용 할 시에 문제발생 제거
plt.rcParams["axes.unicode_minus"] = False
# title 사이즈 지정
plt.rcParams["font.size"] = 16
# x, y의 tick들에 대한 크기도 지정 가능
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 8
# label의 크기 지정
plt.rcParams["axes.labelsize"] = 14

font_path = "C:/Windows/Fonts/malgun.ttf"
# FontProperties.get_name()으로 font family 객체 생성
font_family = fm.FontProperties(fname=font_path).get_name()
# font family 객체를 matplotlib설정에 사용
plt.rcParams["font.family"] = font_family

x_data = range(50)
y_data = np.random.randint(-100,100,50).cumsum() # -100~100사이 50개 난수 생성
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")
plt.plot(x_data, y_data,"r") # "r"은 red 색 지정
plt.show()
```

## 판다스 활용
1. Series 활용
2. DataFrame 활용

### 시리즈를 사용한 그래프
기본적으로 index가 x값으로 사용된다!

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

s = pd.Series(np.random.randn(10),
             index=np.arange(0,100,10))
plt.plot(s, label="오렌지가격") # label로 범례를 표현!
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")
plt.legend() # legend()를 사용해야 범례가 보인다.
plt.show()
```

### 데이터프레임을 사용한 그래프

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,4)

df = pd.DataFrame(np.random.randint(50,100,(10,3)),
                 columns=["이지안","박동훈","홍길동"],
                 index=np.arange(0,10,1))

plt.plot(df) # label로 범례를 표현!
plt.title("가격변동 추이")
plt.xlabel("기간(일)")
plt.ylabel("가격")
plt.show()
```

## bar차트 그리기
1. 기본 형식 < bar, barh
2. 그래프 분석 방식

### 기본 bar 그래프

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,4)
plt.rcParams["axes.grid"] = False

s = pd.Series(np.random.randint(50,100,(3,)),
                 index=["국어","영어","수학"])

s.plot(kind="bar", label="각 과목 성적")
# plt.plot(s)와 동일하다!
# kind로 그래프형식을 지정 ="bar", ="barh" < vertical, horizontal
plt.legend()
plt.show()
```

### bar 그래프 분석 방식

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# 3명 데이터: 4과목에 대한 데이터
df = pd.DataFrame(np.random.randint(50,100,(3,4)),
                 columns=["국어","영어","수학","물리"],
                 index=["이지안","박동훈","홍길동"])

df2 = pd.DataFrame(np.random.randint(50,100,(4,3)),
                 index=["국어","영어","수학","물리"],
                 columns=["이지안","박동훈","홍길동"])
df.plot(kind="bar") # 사람별 잘 하는 과목, 못 하는 과목 비교
df2.plot(kind="bar") # 과목별 3사람의 점수 비교
plt.show()
```

## scatter(산점도) 사용

```python
%matplotlib inline
"""
산점도: 두개의 독립변수 x,y의 관계를 알아보기 위해 사용
"""
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.random.normal(1,1,(100,1)) # 평균1,편차1,사이즈(100,1)
y = np.random.normal(-2,4,(100,1)) # 평균-2,편차4,사이즈(100,1)

plt.scatter(x,y)
plt.show()
# 산점도를 사용해서 경향성을 확인 가능
```

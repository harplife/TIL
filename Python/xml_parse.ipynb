# Read XML file

```python
from xml.dom import minidom


mydoc = minidom.parse(r'C:\python_ML\find_size\save_annot\num_10.xml')

print(mydoc.toprettyxml())
```

```python
filename = mydoc.getElementsByTagName('filename')[0].childNodes[0].data

filepath = mydoc.getElementsByTagName('path')[0].childNodes[0].data

print(filename)
print(filepath)
```

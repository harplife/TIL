## pymysql로 DB 연결

    database로부터 데이터를 읽어서 데이터프레임 생성
    mysql database 사용
    pymysql 모듈 사용 (conda install pymysql)
    
    mysql, oracle, DB2, infomix, cybase;
    데이터베이스라 부르긴 하지만 실제론 DBMS (Database Management System)이라 칭한다.
    얘내들이 데이터베이스인게 아니라 데이터베이스 관리 소프트웨어인 겸이다.
    
    1. pymysql로 데이터베이스 연결
    2. 판다스.read_sql()로 sql구문 실행

```python
import pymysql.cursors
import numpy as np
import pandas as pd

'''
mysql library 데이터베이스/스키마랑 연결
'''
conn = pymysql.connect(host="localhost", # mysql 현재 사용되는 ip주소
                       # 원래는 포트 번호도 지정되야 되지만 일단 skip
                       user="python",
                       password="python",
                       db="library", # 사용할 db
                       charset="utf8") # 한국어.. utf8 지정 꼭!

# library에 books 테이블에 있는 데이터:
# bisbn, btitle, bauthor, bprice, ...

keyword = input("키워드를 입력하세요: ") # 일단 검색할 키워드
sql = "SELECT bisbn,btitle,bauthor,bprice \
        FROM books WHERE btitle LIKE '%{}%'".format(keyword) # 사용할 sql구문 # sql키워드들은 대문자 쓰는게 원칙
'''
SELECT 열(들) FROM 테이블 WHERE 지정 열 LIKE %키워드%
* LIKE 구문은 뭔가를 포함한다~ 라는 뜻
** '%문자열%' quatation해줘야 되네~
'''

# sql실행 결과를 dataframe으로 생성
# dataframe = pd.read_sql(sql구문, con=db접속자)
df = pd.read_sql(sql, con=conn)
display(df)
conn.close()
```

## Pandas로 DB 데이터 .json파일로 저장

```python
import pymysql.cursors
import numpy as np
import pandas as pd

conn = pymysql.connect(host="localhost",
                       user="python",
                       password="python",
                       db="library",
                       charset="utf8")

keyword = input("키워드를 입력하세요: ")
sql = "SELECT bisbn,btitle,bauthor,bprice \
        FROM books WHERE btitle LIKE '%{}%'".format(keyword)

df = pd.read_sql(sql, con=conn)
display(df)
conn.close()

# 판다스의 .to_json(파일경로\파일이름.json) 함수를 사용한다.
df.to_json("./data/sample/books.json")
print("file saved!")
```

## 저장한 .json 파일 읽기

### 파일 open 후 판다스로 읽기

```python
%%time

"""
일단 파이썬의 파일처리기능을 이용하여 file을 열고 데이터를 읽음
파이썬 내부 함수 open("파일명",읽는방식)
"""
import pandas as pd
import json

file = open("./data/sample/books.json","r") # r = read-only

my_dict = json.load(file)

file.close()

df = pd.DataFrame(my_dict)
display(df)
```

### 판다스로 파일 열고 읽기

```python
%%time

"""
걍 판다스를 사용해서 json파일 읽고 출력
비교결과 파이썬 file이용해서 json파일 읽는게 쫌 더 빠르다.
"""
import pandas as pd

df = pd.read_json("./data/sample/books.json")
display(df)
```

## urllib, json으로 로드하고 판다스로 읽기

```python
%%time

"""
파이썬 내부 패키지 사용해서 open api json파일 read and print
"""
import numpy as np
import pandas as pd
import urllib
import json

movie_url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=430156241533f1d058c603178cc3ca0e&targetDt=20190219"

movie_page = urllib.request.urlopen(movie_url) # urllib 사용해서 URL 클래스 생성
json_page = json.loads(movie_page.read()) # 생성된 URL클래스 .read()함수로 읽기

df = pd.DataFrame(json_page)

display(df)
```

```python
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'C:/Users/zion/Music/Starcraft_sounds/Terran/SCV/EDrRep00.wav'

with wave.open(filename, 'r') as signal_wave:
    sample_rate = 352000
    data = np.frombuffer(signal_wave.readframes(sample_rate), dtype=np.int16)
    left, right = data[0::2], data[1::2]    
    lf, rf = abs(np.fft.rfft(left)), abs(np.fft.rfft(right))
```

```python
plt.figure(1)

plot_a = plt.subplot(211)
plot_a.plot(data)
plot_a.set_xlabel('Frames')
plot_a.set_ylabel('Power')

plot_b = plt.subplot(212)
plot_b.specgram(data, NFFT=1024, Fs=sample_rate, noverlap=900)
plot_b.set_xlabel('Time')
plot_b.set_ylabel('Frequency')
plt.show()
```
